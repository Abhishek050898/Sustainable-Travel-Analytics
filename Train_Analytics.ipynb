```python
import pandas as pd
df = pd.read_csv('cleaned_traindata.csv')
df.info()
```

```python
df = df[(df['Origin'] != 'Unknown') & (df['Destination'] != 'Unknown')]
df = df[df['Origin'] != df['Destination']]
df.info()
```

```python
df.drop(['Point of Sale Country', 'Company Name','Unit of Measurement', 'Currency','Department','In Policy','Booking Method','Employee ID'], axis=1, inplace=True)
df['Travel Start Date'] = pd.to_datetime(df['Travel Start Date'])
df['Travel End Date'] = pd.to_datetime(df['Travel End Date'])
df.info()
```

```python
def check_route(row):
    cities = ['Calais', 'Paris', 'Brussels', 'Ebbsfleet International Station', 'Strasbourg', 'Frankfurt', 'Dax', 'Gare Du Nord']
    for city in cities:
        if city in row['Route']:
            return True
    return False

df['nonuk'] = df.apply(check_route, axis=1)
print(df[df['nonuk'] == True])
```

```python
df.to_csv('Train_cleaned.csv', index=False)
```

```python
# Load the datasets
train_data = pd.read_csv('train_data_with_cities.csv')
air_data = pd.read_excel('AirTicket Data.xlsx')
```

```python
origin_unique_air = set(air_data['Origin City'])
destination_unique_air = set(air_data['Destination City'])
all_unique_places_air = list(origin_unique_air.union(destination_unique_air))
print(all_unique_places_air)
```

```python
# Rename columns in train_data to match the desired final dataset
train_data = train_data.rename(columns={
    'Mileage': 'Distance',
    'Carbon Emission': 'CO2 Emission',
    'Ticket Expense': 'Air Expense'
})

# Add 'Mode' column
train_data['Mode'] = 'Train'
air_data['Mode'] = 'Air'

# Concatenate datasets
combined_data = pd.concat([train_data, air_data], ignore_index=True)

# Find combinations of 'Origin City' and 'Destination City' that have both modes
city_combinations = combined_data.groupby(['Origin City', 'Destination City']).filter(lambda x: set(x['Mode']) == {'Train', 'Air'})

# Now, we filter out only those rows that belong to these combinations
final_data = combined_data[
    (combined_data['Origin City'].isin(city_combinations['Origin City'])) & 
    (combined_data['Destination City'].isin(city_combinations['Destination City']))
]

# Select relevant columns
selected_columns = [
    'Travel Start Date', 'Travel End Date', 'Origin City', 'Destination City',
    'Segment Count', 'Distance', 'CO2 Emission', 'Air Expense', 'Mode'
]
final_data = final_data[selected_columns]

# Save to Excel
final_data.to_csv('matched_travel_data.csv', index=False)
```
